### Ingeniería de Datos con Pandas (Limpieza de datos)

```python
#Importando las librerias a utilizar
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
#cargar data set con parseo de fechas y encoding
df = pd.read_csv('../data/ventas_techventas.csv', parse_dates=['fecha'], encoding='Latin-1')
print(df.info())
```

- **Observaciones**

En base a la información mostrada, se puede deducir que el dataset tiene 60 filas y 11 columnas, de las cuales 3 son de tipo decimal, 7 de ellos tipo cadena de texto y uno de formato de fechas. 9 de las columnas tienen 49 campos que no son nulos, 1 de esas columnas tiene 48 campos que no son nulos y solo una columna cuenta con los 60 campos completos sin ser nulos.

```python
#Identificando los campos nulos
df.isna()
```

```python
#Identificando el total de campos nulos por columnas
df.isna().sum()
```

```python
#Viendo el porcentaje de los valores nulos
(df.isna().sum() / len(df)) * 100
```

```python
#Creando copia del dataframe
df2 = df.copy()
```

```python
#Imputando los datos nulos de las columnas
df2['cliente_id'] = df2['cliente_id'].fillna('Desconocido')
df2['producto'] = df2['producto'].fillna('Desconocido')
df2['fecha'] = df2['fecha'].ffill()
df2['categoria'] = df2['categoria'].fillna('Desconocido')
df2['region'] = df2['region'].fillna('Desconocido')
df2['cantidad'] = df2['cantidad'].fillna(0)
df2['precio_unitario'] = df2['precio_unitario'].fillna(0)
df2['descuento'] = df2['descuento'].fillna(0)
print(df2)
```
