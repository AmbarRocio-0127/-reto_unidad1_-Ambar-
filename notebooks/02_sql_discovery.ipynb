### Exploración de datos con SQL (Data Recovery)

```python
#importando las librerias a utilizar
import sqlite3
import pandas as pd
import sqlalchemy as alchemy
```

```python
#Importación del dataframe con los datos sucios
dataframe = pd.read_csv("../data/ventas_techventas.csv", encoding="latin-1")
```

```python
#Se crea el motor de la BD con la ruta donde ha de crearse la base de datos
motor_db = alchemy.create_engine("sqlite:///../output/ventas_techventas.db")
```

```python
#Se crea la tabla de ventas en la BD al exportar el dataframe
dataframe.to_sql("ventas_techventas", motor_db, if_exists="replace", index=False)
```
