### Exploración de datos con SQL (Data Recovery)

```python
#importando las librerias a utilizar
import sqlite3
import pandas as pd
import sqlalchemy as alchemy
```

```python
#Importación del dataframe con los datos sucios
dataframe = pd.read_csv("../data/ventas_techventas.csv", encoding="latin-1")
```

```python
#Se crea el motor de la BD con la ruta donde ha de crearse la base de datos
motor_db = alchemy.create_engine("sqlite:///../output/ventas_techventas.db")
```

```python
#Se crea la tabla de ventas en la BD al exportar el dataframe
dataframe.to_sql("ventas_techventas", motor_db, if_exists="replace", index=False)
```

```python
#Explorando las columnas de la tabla ventas_techventas
df = pd.read_sql("""
PRAGMA table_info(ventas_techventas);
""", motor_db)

print(df)
```

```python
#Ya creadas, Se accede a las consultas de la BD en modo lectura y se guardan en la lista llamada consulta dividiendo 
#las seis consultas en bloques por los comentarios (--)
with open("../sql/02_sql_discovery.sql","r", encoding="latin-1") as sql:
    discovery = sql.read()
consultas = discovery.split("-- ")
```

```python
#Se recorren los bloques correspondientes de cada consulta, se guardan en un diccionario el nombre de la consulta 
# y el codigo sql para poder acceder a cada una de manera individual.

diccionario_consultas = {}
for consulta in consultas:
    if consulta.strip():
        nombre, sql = consulta.split("\n",1)
        diccionario_consultas[nombre.strip()] = sql.strip()
```
