```python
#importando las librerias correspondientes

import numpy as np
import pandas as pd
```

```python
#cargando los datos de csv y creando los arrays de las cantidades y de los precios
#y accediendo a ellos por medio de un ciclo for
datos = np.genfromtxt("../data/ventas_techventas.csv", encoding="latin-1", delimiter=",", skip_header=1)
cantidad = datos[:,7]
precio = datos[:,8]

print("Cantidad\tPrecio")

for i in range(len(cantidad)):
    print(f"{cantidad[i]}\t\t{precio[i]}")
```

- **Explorando los datos del dataset**

```python
print("---Resumen del Dataframe---\n")
dataframe.info()
print("\n---Resumen de Estadística descriptiva en cuanto a las columnas numericas del dataframe---\n")
dataframe.describe()
```

```python
#visualizando las ultimas filas del data set y verificando la suma total de nulos 
print("---Suma total de Valores Nulos---\n")
print(dataframe.isnull().sum(),"\n")
print("---Ultimas 4 filas del dataset---\n")
print(dataframe.tail(4))
```

```python
#Realizando el calculo del producto punto tomando los valores de cantidad y precio 
# y la sumatoria de total de las columnas
ingreso_bruto = cantidad @ precio
print("Comparando valores (Sin limpiar los datos)"),
print("\nCantidad total de productos vendidos: ",cantidad.sum())
print("\nSumatoria del precio Total de todos los productos: ",precio.sum())
print("\nIngreso Bruto: ",ingreso_bruto)
```

**1.3 Estadísticas con Numpy**

```python
media = np.mean(precio)
mediana = np.median(precio)
desviacion_estandar = np.std(precio)
percent = np.percentile(precio,[25,75])
print("Media de los precios: ", media)
print("\nMediana de los precios: ", mediana)
print("\nDesviación estándar de los precios: ", round(desviacion_estandar,2))
print("\nPercentiles 25 y 75: ", percent)
```

**Análisis**

- En base a las operaciones realizadas se puede observar que el promedio del precio de la cantidad de productos vendidos corresponde a 378.25 dólares durante los ultimos 6 meses. La mediana de las ventas corresponde a 120.0 dolares, lo que se podria interpretar en que al menos la mitad de las ventas oscila alrededor de este valor (120). En comparación con el promedio, la desviación estándar es de 445.06 dolares, de lo que se podría decir que el valor de las ventas varia del promedio con una cantidad de 66.81 (diferencia de la desviación estándar y el promedio) y, los percentiles que presentan los valores de 92.5 y 350, lo que significa que el 25% de las ventas esta debajo de 92.5 y el 75% esta por encima de 350.
